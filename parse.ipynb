```python
import os
import json
from pathlib import Path

rootdir = "/Users/eashan22/Desktop/Internship 2021/bbrv2/Brian's Project/bost-dtn"
fileList = []
c = 0

for root, directories, files in os.walk(rootdir):
    for file in files:
        if file.startswith("ss") and file.endswith("json"):
            # print(os.path.join(root,file))
            fileList.append(os.path.join(root,file))
            c+=1

print(f"Total files: {c}")
```

```python
cubic, bbr2, bbr = [], [], []
key1, key2, key3 = 'cubic_data_segs', 'bbr2_data_segs', 'bbr_data_segs'

for i,f in enumerate(fileList):
    try:
        path = Path(f)
        # The MongoDB JSON dump has one object per line, so this works for me.
        data = [json.loads(line) for line in open(f, 'r')]
        
        # i = len(data)-2
        for i in range(len(data)):
            if key1 in data[i].keys():
                if data[i]['cubic_data_segs']!=0:
                    cubic.append(( path.parent.name, data[i]['cubic_data_segs']) )
                elif data[i]['bbr2_data_segs']!=0:
                    bbr2.append(( path.parent.name, data[i]['bbr2_data_segs']) )
                elif data[i]['bbr_data_segs']!=0:
                    bbr.append(( path.parent.name, data[i]['bbr_data_segs']) )

    except:
        print(f"Index: {i}, Parent: {path.parent.name}")

print(len(cubic), len(bbr2), len(bbr))

# cubic, bbr2, bbr = [], [], []

# for i,f in enumerate(fileList):
#     try:
#         path = Path(f)
#         # The MongoDB JSON dump has one object per line, so this works for me.
#         data = [json.loads(line) for line in open(f, 'r')]
#         index = len(data)-2
#         if data[index]['cubic_data_segs']!=0:
#             cubic.append(( path.parent.name, data[index]['cubic_data_segs']) )

#         elif data[index]['bbr2_data_segs']!=0:
#             bbr2.append( (path.parent.name, data[index]['bbr2_data_segs']) )

#         elif data[index]['bbr_data_segs']!=0:
#             bbr.append( (path.parent.name, data[index]['bbr_data_segs']) )

#         # print( f"[{i}]. cubic:{data[index]['cubic_data_segs'] : >10} | bbr2:{data[index]['bbr2_data_segs'] : ^15} | bbr:{data[index]['bbr_data_segs'] : >10}" )
#     except:
#         print(f"Index: {i}, Parent: {path.parent.name}")
#         # data = [json.loads(line) for line in open(f, 'r')]
#         # index = len(data)-2
#         # print( f"[{i}]. cubic:{data[index]['cubic_data_segs'] : >10} | bbr2:{data[index]['bbr2_data_segs'] : >10}" ) 
#     # else:
#         # print(f"File {i}: Error in file: {f}")
```
